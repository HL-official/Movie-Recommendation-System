```python
import numpy as np
import pandas as pd
import ast
```

```python
movies  = pd.read_csv('movies.csv')
```

```python
movies.head()
```

```python
movies.head(1)
```

```python
movies.info()
```

```python
movies = movies[['id','title','genres','overview','credits','keywords']]
```

```python
movies.info()
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].overview
```

```python
movies['genres']=movies['genres'].apply(lambda x: x.replace(" ",""))
movies['genres']=movies['genres'].apply(lambda x: x.replace("-"," "))
movies['credits']=movies['credits'].apply(lambda x: x.replace(" ",""))
movies['credits']=movies['credits'].apply(lambda x: x.replace("-"," "))
movies['keywords']=movies['keywords'].apply(lambda x: x.replace(" ",""))
movies['keywords']=movies['keywords'].apply(lambda x: x.replace("-"," "))
```

```python
movies['genres']=movies['genres'].apply(lambda x:x.split())
movies['credits']=movies['credits'].apply(lambda x:x.split())
movies['keywords']=movies['keywords'].apply(lambda x:x.split())
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['overview']=movies['overview'].apply(lambda x: [i.replace("-","") for i in x])
movies['overview']=movies['overview'].apply(lambda x: [i.replace("'","") for i in x])
```

```python
movies.head()
```

```python
def convert5(obj):
    L=[]
    counter = 0
    for i in obj:
        if counter !=5:
            L.append(i)
            counter+=1
        else:
            break

    return L
```

```python
movies['credits']=movies['credits'].apply(convert5)
```

```python
movies['credits'][0]
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['credits']
```

```python
movies.head()
```

```python
new_df=movies[['id','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
new_df['tags'][0]
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))

    return ' '.join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```
